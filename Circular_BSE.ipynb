```python
from requests_html import HTMLSession
from bs4 import BeautifulSoup as bs
import pandas as pd
import gspread
import datetime
```

```python
s = HTMLSession()
header = {
'scheme':'https',
'Accept':'application/json, text/plain, */*',
'Accept-Language':'en-US,en;q=0.8',
'Referer':'https://www.bseindia.com/',
'Sec-Ch-Ua':'"Not_A Brand";v="8", "Chromium";v="120", "Brave";v="120"',
'Sec-Ch-Ua-Mobile': '?0',
'Sec-Ch-Ua-Platform':'"Windows"',
'Sec-Fetch-Mode':'cors',
'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36'
}
```

```python
Date = datetime.datetime.now().strftime('%Y%m%d')
```

```python
All_data = []
count = 1
while True:
   url = f'https://api.bseindia.com/BseIndiaAPI/api/AnnSubCategoryGetData/w?pageno={count}&strCat=-1&strPrevDate={Date}&strScrip=&strSearch=P&strToDate={Date}&strType=C&subcategory='
   r = s.get(url,headers=header) 
   if 'Table' in  r.json() and r.json()['Table']:
        data = r.json()['Table']
        temp_df = pd.DataFrame(data)
        All_data.append(temp_df)
        print('Page.no =',count)
        count += 1  
   else:
        break
```

```python
gc = gspread.service_account(filename='creds.json')
spreadsheet_name = 'Notifications and Listings'
sheet_name = 'BSE_Circulars'
sh = gc.open(spreadsheet_name).worksheet(sheet_name)
```

```python
Final_Data = pd.concat(All_data,ignore_index=True)
Final_Data.head()
filtered_df = Final_Data[['SCRIP_CD','NEWSSUB','DT_TM','SLONGNAME','SUBCATNAME','ATTACHMENTNAME']]
filtered_df.head(20)
```
