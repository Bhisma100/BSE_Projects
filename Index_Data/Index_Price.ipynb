```python
from selenium import webdriver
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.common.by import By
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.keys import Keys
from bs4 import BeautifulSoup as bs
import pandas as pd
import time
```

```python
dfs = []
```

```python
driver = webdriver.Chrome()
url = 'https://www.bseindia.com/markets/keystatics/Keystat_index.aspx'
driver.get(url)
```

```python
Click_buttons = []
for x in range(0,6):
    link = f'//*[@id="ContentPlaceHolder1_gvReport_Linkbtn_{x}"]'
    Click_buttons.append(link)
```

```python
for link in Click_buttons:
    try:
        # Wait for the element to be clickable
        element = WebDriverWait(driver, 10).until(
            EC.element_to_be_clickable((By.XPATH, f'{link}'))
        )

        # Scroll the element into view
        driver.execute_script("arguments[0].scrollIntoView();", element)

        # Click the element
        element.click()
        time.sleep(5)  # Give some time for the page to load

        # Continue with your scraping logic
        page_source = driver.page_source
        soup = bs(page_source, 'html.parser')
        table = soup.find('table', id='ContentPlaceHolder1_gvYearwise')
        rows = table.find_all('tr')
        headers = [th.text.strip() for th in rows[0].find_all('th')]
        data = []
        for row in rows[1:]:
            row_data = [td.text.strip() for td in row.find_all('td')]
            data.append(row_data)
        temp_df = pd.DataFrame(data, columns=headers)
        dfs.append(temp_df)

        # Scroll to the top of the page
        driver.execute_script("window.scrollTo(0, 0);")

    except Exception as e:
        print(f"Error processing element {i}: {e}")

# Concatenate all DataFrames into a single DataFrame
final_df = pd.concat(dfs, ignore_index=True)

# Close the WebDriver
driver.quit()
```
