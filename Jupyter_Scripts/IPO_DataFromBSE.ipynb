```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import Select
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.support.ui import WebDriverWait
from requests_html import HTMLSession
from bs4 import BeautifulSoup as bs
import pandas as pd
```

```python
driver = webdriver.Chrome()
driver.get('https://www.bseindia.com/markets/PublicIssues/IPOIssues_new.aspx?id=2&Type=P')
dropdown = driver.find_element(By.ID, "ctl00_ContentPlaceHolder1_ddlIssueType")
select = Select(dropdown)   
```

```python
select.select_by_visible_text('Public Issue-Fixed Price')
submit_button = driver.find_element(By.ID, "ctl00_ContentPlaceHolder1_btnSubmit")
submit_button.click()   
WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.ID, "ctl00_ContentPlaceHolder1_tblID")))
table_data = driver.find_element(By.ID, "ctl00_ContentPlaceHolder1_tblID")
table_html = table_data.get_attribute('outerHTML')
```

```python
soup = bs(table_html, 'html.parser')
```

```python
Name_Links = []
for links in soup.find_all('a'):
    Data = {
        'Company_Name':links.text,
        'link' : f"https://www.bseindia.com{links.get('href')}"
    }
    Name_Links.append(Data)
    
```

```python
header ={
'scheme':'https',
'Accept':'application/json, text/plain, */*',
'Accept-Language':'en-US,en;q=0.8',
'Referer':'https://www.bseindia.com/markets/PublicIssues/IPOIssues_new.aspx?id=2&Type=P',
'Sec-Ch-Ua':'"Not_A Brand";v="8", "Chromium";v="120", "Brave";v="120"',
'Sec-Ch-Ua-Mobile': '?0',
'Sec-Ch-Ua-Platform':'"Windows"',
'Sec-Fetch-Mode':'cors',
'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36'
}
```

```python
IPO_Data = []    
for i in range(300):
    test_link = Name_Links[i]['link']
    print(f'Link{i}')
    s = HTMLSession()
    response = s.get(test_link)
    soup1 = bs(response.text,'html.parser')
    div_element = soup1.find('div',id = 'ctl00_ContentPlaceHolder1_UpdatePanel1')
    table = div_element.find('table').find_all('tr')
    for rows in table[4:17]:
        row = rows.find_all('td')
        if 'Security Type' in row[0].text:
            Dictionary1 = {row[0].text: row[-1].text}
        elif 'Symbol' in row[0].text:
            Dictionary1 = {row[0].text: row[-1].text}
        elif 'Issue Period' in row[0].text:
            Dictionary1 = {row[0].text: row[-1].text}
        elif 'Issue Size  No. of Shares' in row[0].text:
            Dictionary1 = {row[0].text: row[-1].text}
        elif 'Price Band' in row[0].text:
            Dictionary1 = {row[0].text: row[-1].text}
        elif 'Face Value' in row[0].text:
            Dictionary1 = {row[0].text: row[-1].text}
        elif 'Tick Size' in row[0].text:
            Dictionary1 = {row[0].text: row[-1].text}
        elif 'Market Lot' in row[0].text:
            Dictionary1 = {row[0].text: row[-1].text}
        elif 'Minimum Bid Quantity' in row[0].text:
            Dictionary1 = {row[0].text: row[-1].text}
        else:
            Dictionary1 = {row[0].text: 'NA'}
        print(Dictionary1)
        IPO_Data.append(Dictionary1)
    
    # Append the dictionary to the IPO_Data list
    
```

```python
driver.close()
```

```python
Data = []
for i in range(0, len(IPO_Data), 13):
    data = {}
    for entry in IPO_Data[i:i+13]:
        for key, value in entry.items():
            data[key] = value
    Data.append(data)
```

```python
for i in Data[0:len(Data)-1]:
    print(i)
```

```python
df1 = pd.DataFrame(Data[0:len(Data)-1])
df1.head(50)
```
