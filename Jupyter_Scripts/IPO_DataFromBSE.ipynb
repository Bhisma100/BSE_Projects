```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import Select
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.support.ui import WebDriverWait
from requests_html import HTMLSession
from bs4 import BeautifulSoup as bs
import pandas as pd
import time
```

```python
driver = webdriver.Chrome()
driver.get('https://www.bseindia.com/markets/PublicIssues/IPOIssues_new.aspx?id=2&Type=P')
dropdown = driver.find_element(By.ID, "ctl00_ContentPlaceHolder1_ddlIssueType")
select = Select(dropdown)
select.select_by_visible_text('Public Issue-Book Building')
submit_button = driver.find_element(By.ID, "ctl00_ContentPlaceHolder1_btnSubmit")
submit_button.click()   
WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.ID, "ctl00_ContentPlaceHolder1_tblID")))
table_data = driver.find_element(By.ID, "ctl00_ContentPlaceHolder1_tblID")
table_html = table_data.get_attribute('outerHTML')  
```

```python
soup = bs(table_html, 'html.parser')
```

```python
Name_Links = []
for links in soup.find_all('a'):
    Data = {
        'Company_Name':links.text,
        'link' : f"https://www.bseindia.com{links.get('href')}"
    }
    Name_Links.append(Data)
for i in range(0,1):
    url1 = Name_Links[i]['link']
    
```

```python
Ten_year = []
IPOCodes = []
for i in range(0,len(Name_Links)):
        year = int(Name_Links[i]['link'][-4:])
        if year > 2012 :
            Ten_year.append(Name_Links[i]['link'])
for Links in Ten_year:
      code = Links.split('&')
      if len(code[4]) < 10:
        IPOCodes.append(code[4][-3:])
      else:
        IPOCodes.append(code[4][-4:])
```

```python
print(Links.split('&'))
```

```python
for k in IPOCodes:
    print(k)
```

```python
driver1 = webdriver.Chrome()
Subscription_Ratio = []
for m in range(0,len(IPOCodes)):
    print(IPOCodes[m])
    url = f'https://www.bseindia.com/markets/publicIssues/CummDemandSchedule.aspx?ID={IPOCodes[m]}&status=L'
    driver1.get(url)
    table2 = 1
    try:
        Data = WebDriverWait(driver1,5).until(EC.presence_of_element_located((By.XPATH,'//*[@id="divID"]/table/tbody/tr/td/table/tbody'))).text
        # print(Data)
        lines = Data.split('\n')
        dict = {'Name': lines[1]}
        Subscription_Ratio.append(dict)
        retail_found = False
        total_found = False
        for line in lines:
            if 'Retail Individual Investors (RIIs)' in line:
                dict = {'Retail Individual Investors (RIIs)' : f'{line[len('Retail Individual Investors(RIIs)')+3:]}'}
                Subscription_Ratio.append(dict)
                retail_found = True
            if 'Total' in line:
                dict = {'Total':f'{line[len('Total')+1:]}'}
                Subscription_Ratio.append(dict)
                total_found = True
        if not retail_found:
            dict = {'Retail Individual Investors (RIIs)': 'NA'}
            Subscription_Ratio.append(dict)
        if not total_found:
                dict = {'Total': 'NA'}
                Subscription_Ratio.append(dict)

    except :
        print('No Table1')
        table2 = 0
    if table2 == 0:
        try:
            Stlong = WebDriverWait(driver1,10).until(EC.presence_of_element_located((By.XPATH,'//*[@id="divID"]/table[1]/tbody'))).text
            Names = Stlong.split('\n')
            dict = {'Name':Names[0]}
            Subscription_Ratio.append(dict)
            Data = WebDriverWait(driver1,5).until(EC.presence_of_element_located((By.XPATH,'//*[@id="divID"]/table[2]/tbody'))).text
            lines = Data.split('\n')
            retail_found = False
            total_found = False
            for line in lines:
                if 'Retail Individual Investors(RIIs)' in line:
                    dict = {'Retail Individual Investors (RIIs)' : f'{line[len('Retail Individual Investors(RIIs)')+3:]}'}
                    Subscription_Ratio.append(dict)
                    retail_found = True
                if 'Total' in line:
                    dict = {'Total':f'{line[len('Total')+1:]}'}
                    Subscription_Ratio.append(dict)
                    total_found = True
            if not retail_found:
                dict = {'Retail Individual Investors (RIIs)': 'NA'}
                Subscription_Ratio.append(dict)
            if not total_found:
                dict = {'Total': 'NA'}
                Subscription_Ratio.append(dict)
        except:
            print('No Table2')
    time.sleep(1)
driver1.close()
```

```python
for h in Subscription_Ratio:
    print(h)
```

```python
Data = []
for i in range(0, len(Subscription_Ratio), 3):
    data = {}
    for entry in Subscription_Ratio[i:i+3]:
        for key, value in entry.items():
            data[key] = value
    Data.append(data)
```
