```python
from requests_html import HTMLSession
from bs4 import BeautifulSoup as bs
import pandas as pd
import gspread
import datetime
```

```python
s = HTMLSession()
header = {
'scheme':'https',
'Accept':'application/json, text/plain, */*',
'Accept-Language':'en-US,en;q=0.8',
'Referer':'https://www.bseindia.com/',
'Sec-Ch-Ua':'"Not_A Brand";v="8", "Chromium";v="120", "Brave";v="120"',
'Sec-Ch-Ua-Mobile': '?0',
'Sec-Ch-Ua-Platform':'"Windows"',
'Sec-Fetch-Mode':'cors',
'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36'
}
```

```python
start_date = datetime.date(2023, 9, 1)
end_date = datetime.date.today()
```

```python

All_data = []
current_date = start_date
while current_date <= end_date:
    first_day = current_date.replace(day=1)
    last_day = (datetime.date(current_date.year, current_date.month % 12 + 1, 1) - datetime.timedelta(days=1))
    date_from = first_day.strftime('%Y%m%d')
    date_to = last_day.strftime('%Y%m%d')
    print(f"{date_from} to {date_to}")
    count = 1
    while True:
          url = f'https://api.bseindia.com/BseIndiaAPI/api/AnnSubCategoryGetData/w?pageno={count}&strCat=Company+Update&strPrevDate={date_from}&strScrip=&strSearch=P&strToDate={date_to}&strType=C&subcategory=-1'
          r = s.get(url,headers=header) 
          if 'Table' in  r.json() and r.json()['Table']:
               data = r.json()['Table']
               temp_df = pd.DataFrame(data)
               All_data.append(temp_df)
               print('Page.no =',count)
               count += 1  
          else:
               break
    current_date = datetime.date(current_date.year + current_date.month // 12, (current_date.month % 12) + 1, 1)     
```

```python
Final_Data = pd.concat(All_data,ignore_index=True)
Final_Data.head(20)
```

```python
Final_DataInv = Final_Data[Final_Data['HEADLINE'].str.contains('Earning Call|Earning Call Transcript|Earning Conference Call|Con-Call|Con-call|transcript|Transcript of earning|Audio recording|Audio Recording|recording |audio',case= False) | Final_Data['SUBCATNAME'].str.contains('Earnings Call Transcript',case= False)|Final_Data['MORE'].str.contains('Earning Call|Earning Call Transcript|Earning Conference Call|Con-Call|Con-call|transcript|Transcript of earning|Audio recording|Audio Recording|recording |audio',case= False)|Final_Data['NEWSSUB'].str.contains('Earning Call|Earning Call Transcript|Earning Conference Call|Con-Call|Con-call|transcript|Transcript of earning|Audio recording|Audio Recording|recording |audio',case= False)]
```
